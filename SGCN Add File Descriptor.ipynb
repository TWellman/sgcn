This simple script adds an explicit file title to the existing items in the SGCN source data repository indicating the explicit file that should be processed as source data for the state and year. This is not a very sophisticated approach and should only be run one time to clean up the current set.

The script adds the text that Abby came up with, "Process ready version of original file," to just the tab delimited text file that contains the raw submission from the state. With this in place, we don't have to make an assumption about which file to process into the final integrated SGCN database. There is also a convention for titling updates and the original submission files that will be applied individually as the source data are managed in ScienceBase.

```python
# Get dependencies and set up an sb session
import re
import requests
import pysb
import os
sb = pysb.SbSession()
username = input("Username: ")
sb.loginc(str(username))
```

```python
# Get all items from the collection for SGCN
r = requests.get("https://www.sciencebase.gov/catalog/items?parentId=56d720ece4b015c306f442d5&format=json&fields=title,files&max=100").json()    
```

```python
# Loop through the items
for item in r['items']:
    newItem = {}
    newItem["id"] = item["id"]
    dateCollected = re.findall(r'\d+', item['title'])[0]
    if dateCollected == "2015":
        newFiles = []
        for file in item["files"]:
            if file['name'][-9:] == '_2015.txt':
                file["title"] = "Process ready version of original file"
                newFiles.append(file.copy())
            else:
                newFiles.append(file.copy())
    elif dateCollected == "2005":
        newFiles = []
        for file in item["files"]:
            if file['name'][-25:] == 'Species_Original_List.txt':
                file["title"] = "Process ready version of original file"
                newFiles.append(file.copy())
            else:
                newFiles.append(file.copy())
    newItem["files"] = newFiles
    sb.update_item(newItem)
    print(newItem)
```
