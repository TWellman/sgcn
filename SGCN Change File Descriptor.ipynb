In working through the data management process for SGCNB files, it became obvious that it will be simpler to have a single flag that we are looking for on repository files. This script changes the "Process ready version..." title descriptor on files to simply "Process File" for both updates and original files. This way, the data manager simply needs to flag whatever file is ready to be fully processed into the SGCN database with that title.

We will also be feeding some information back to the ScienceBase Items as part of processing so that the items themselves are "aware" of when they've been processed.

```python
# Get dependencies and set up an sb session
import re
import requests
import pysb
import os

sb = pysb.SbSession()
username = input("Username: ")
sb.loginc(str(username))
```

```python
# Get all items from the collection for SGCN
r = requests.get("https://www.sciencebase.gov/catalog/items?parentId=56d720ece4b015c306f442d5&format=json&fields=title,files&max=100").json()    
```

```python
from IPython.display import display

# Loop through the items
for item in r['items']:
    newItem = {}
    newItem["id"] = item["id"]
    newFiles = []
    for file in item["files"]:
        if file["title"] == "Process ready version of original file" or file["title"] == "Process ready version of update file":
            file["title"] = "Process File"
            newFiles.append(file.copy())
        else:
            newFiles.append(file.copy())
    newItem["files"] = newFiles
    sb.update_item(newItem)
    display(newItem)
```
